# define wrLines() and wrChars() functions
- note that 'tobe' has newline chars(\n), but there are none in 'tobeLines'
- both functions should first remove the file named by tp if it exists, to get a fresh start.
- os.remove will throw an error if the file doesn't exist,
so either catch it or verify the file exists before trying to delete it
    - wrLines writes the contents of 'tobeLines' to a file, using the fd.writelines() method. 
you must add a '\n' explicitly to each line. if you are on windows, use '\n' as your newline character, so the counts come out correctly
    - wrChars writes the contents of 'tobe', one character at a time, using fd.write() method
- after the file write is complete, both functions should get the file byte size,
using os.stat, and return it.

````python
import os

# get a tmp file pathname. all functions below will use it

import tempfile

# does not create the file, just makes a pathname
tp = tempfile.NamedTemporaryFile().name
````

````python
import os
import tempfile
def wrLines(tp, tobeLines):
    if os.path.exists(tp):
        os.remove(tp)
    with open(tp, 'w') as fd:
        fd.writelines([line+'\n' for line in tobeLines])
    return os.stat(tp).st_size
````

````python
def wrChars(tp, tobe):
    if os.path.exists(tp):
        os.remove(tp)
    with open(tp, 'w') as fd:
        fd.write(tobe)
    return os.stat(tp).st_size
````

````python
# evaluated on windows
wrChars(tp, tobe)
````

# define rdAll()
- read the tp file in one fd.read() call (eagar approach)
- return the number of characters read
- if optional search arg supplied, return length of chars read, and index of where search string is found

````python
def rdAll(tp, *arg):
    with open(tp, 'r') as fd:
        f = fd.read()
        if len(arg) > 0:
            return [len(f), f.find(arg[0])]
        return len(f)
````

# define rdLines()
- read the tp file by using fd.readlines() (eager approach)
- with no arg, return the number of characters read
- with optinal search arg, return the index of the search string in a list

````python
def rdLines(tp, *arg):
    with open(tp,'r') as fd:
        f = ''.join(fd.readlines())
        if len(arg) > 0:
            return [f.find(arg[0])]
        return len(f)
````

# define rdChars
- read the tp file one character at a time using fd.read(1) (lazy approach)
- fd.read(1) will return an empty string when the file is exhausted
- with no arg, return the number of characters read
- with optinal search arg, return the index of the search string in a list

````python
def rdChars(tp, *arg):
    with open(tp,'r') as fd:
        readcnt = 0
        flag = 0
        while True:
        # read next character
            x = fd.read(1)
        # returns empty string when finished
            if x == '':
                return readcnt
            readcnt += 1
            
            if (len(arg) > 0):
                # enter if x matches the first character of the argument
                if arg[0][0] == x:
                    for i in range(1,len(arg[0])):
                        # read next character
                        x = fd.read(1)
                        if x == '':
                            return readcnt
                        readcnt += 1
                        # break if does not match with the argument
                        if x != arg[0][i]:
                            flag = 0
                            break
                        flag = 1
                    # return if the argument string is found
                    if flag == 1:
                        return [readcnt - len(arg[0])]
````

````python
# must eval this cell

import urllib.request
import collections
import re
import bs4
import lxml

hamleturl = 'https://courseworks.columbia.edu/access/content/group/'
hamleturl += 'COMSW3101_002_2015_3/data/hamlet.html'
hamleturl

with urllib.request.urlopen(hamleturl) as ef:
    for j in range(40):
        print(next(ef))
````

````python
# The goal is to return a list of the linecnt, total number of 'speeches'(look at the file format), and a dict showing the number of 'speeches' each character gives
import re
from collections import defaultdict
def hamlet():
    with urllib.request.urlopen(hamleturl) as ef:
        # read hamleturl as string
        s = ef.read().decode("utf-8")
        
        # find speeches
        speech = re.findall('<A NAME=speech[0-9]+><b>[a-zA-Z ]+</b>', s)
        
        # format speaker names
        speaker = re.findall('<b>[a-zA-Z ]+</b>',''.join(speech))
        speaker = [re.sub('<b>','',s) for s in speaker]
        speaker = [re.sub('</b>','',s) for s in speaker]
        
        # count
        d = defaultdict(int)
        for item in speaker:
            d[item] += 1 
            
        return [len(s.splitlines()),len(speech),d]
````

- savings will create a pandas DataFrame that shows how money will grow with compound interest, with specified rates and sizes of deposits and interest

````python
from datetime import datetime
import pandas as pd
````

````python
# with the arg defaults shown, this will 
# calculate the money accumulated after 
# 31 days, with weekly deposits of $10, 
# at a 10% interest rate compounded
# on each business day

def savings(start = '2018-03-01', end = '2018-03-31',\
            depositFreq='W', depositAmount=10., 
            compoundFreq='B', interestRate=.1 ):
    # your code here
    # add dates as index
    dates = pd.date_range(start = start, end = end)
    money = pd.DataFrame(index = dates)
    
    # add day column
    day = [dt.strftime('%A') for dt in dates]
    money['day'] = day
    
    #add deposits column
    deposits = pd.Series(depositAmount, index=pd.date_range(start = start, end = end, freq = depositFreq))
    money['deposits'] = deposits
    
    # add compound column
    compound = pd.Series(interestRate, index=pd.date_range(start = start, end = end, freq = compoundFreq))
    money['compound'] = compound
    
    # handle NA's
    money.fillna(0.0, inplace = True)
    
    # create intser and balser series
    intser = pd.Series(0.0, index=pd.date_range(start = start, end = end, freq = 'D'))
    balser = pd.Series(0.0, index=pd.date_range(start = start, end = end, freq = 'D'))
    
    # computation
    balance = 0
    interest = 0
    for j,t in enumerate(money.itertuples()):
        # pull out row values
        td = t.deposits
        ti = t.compound
        # calculate interest and new balance
        balance += td
        interest = balance * ti
        balance += interest
        # update the series objects
        balser[j] = balance
        intser[j] = interest
    
    # add interest and balance columns
    money['interest'] = intser
    money['balance'] = balser
    
    return money
    
````
